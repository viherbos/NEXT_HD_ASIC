```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import fit_library as fit_lib
%matplotlib nbagg
%reload_ext autoreload
%autoreload 2
```

```python
C_int = 6.17
Low_V = 0.4113
High_V = 1.4335
V_flip = High_V-Low_V
print(V_flip)
```

```python
Q_true = np.array([1.548, 2.678, 4.391, 5.545, 6.721, 10.71, 17.27, 22.22, 28.24, 33, 37.68, 44.12, 
              50.81, 59.2, 67.9, 78.45, 85.48, 92.7, 99.78, 115, 121.8])
Flips  = np.array([0.2359, 0.4140, 0.684, 0.866, 1.0633, 1.6869, 2.7099, 3.4645, 
                  4.3710, 5.0563, 5.7678, 6.6956, 7.6243, 8.7909, 9.9675, 11.2964, 12.1838, 13.064, 
                  13.9885, 15.6967, 16.803])
Q_meas = Flips* V_flip * C_int
```

```python
Q_error = Q_true - Q_meas
```

```python
def error_prototype(x,*param):
    return x**param[0]/param[1]
```

```python
plt.figure()
plt.plot(Flips,Q_error)
plt.plot(Flips,error_prototype(Flips,2.5,60))
```

# CORRECTION

```python
Q_corrected = Q_meas + error_prototype(Flips,2.5,60)
```

```python
plt.figure()
plt.plot(Q_true,(Q_corrected-Q_true)/Q_true*100)
```
